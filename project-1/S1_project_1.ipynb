# 11) in which city  the average detection_drives_count is higher? 

In order to find the average detection_drives_count for eatch city:

first we will count how many items are related to Tel-aviv and how many related Barcelona

after that we  will sum up the values row-wise (up->down the column detection_drives_count)

last we will culculate the average 

```python
tlv_sum_ddc = np.sum(df[df.city.eq(1)]['detection_drives_count'], axis = 0)  
brc_sum_ddc = np.sum(df[df.city.eq(0)]['detection_drives_count'], axis = 0)

tlv_items = df[df.city.eq(1)].shape[0]
brc_items = df[df.city.eq(0)].shape[0]

tlv_avg = tlv_sum_ddc / tlv_items
brc_avg = brc_sum_ddc / brc_items
print('the avg detection_drives_count in Tel-aviv :' + str( tlv_avg))
print('the avg detection_drives_count in Barcelona :' + str( brc_avg))
```

The rate of observations in Tel-aviv is higer then in barcelona
